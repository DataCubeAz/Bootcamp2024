 ## Aggregation with Grouping

| Aggregation| English Description                                   | German Description                                       |
|-------------|-------------------------------------------------------|----------------------------------------------------------|
| count()     | Count the number of non-NA/null entries in each column. | Zhlen der Anzahl von nicht-NA/null Eintrgen in jeder Spalte. |
| sum()       | Calculate the sum of each numeric column.            | Berechnen der Summe jeder numerischen Spalte.            |
| max()       | Find the maximum value in each column.               | Finden des maximalen Werts in jeder Spalte.              |
| min()       | Find the minimum value in each column.               | Finden des minimalen Werts in jeder Spalte.              |
| first()     | Get the first non-NA value in each column.            | Erhalten des ersten nicht-NA-Werts in jeder Spalte.      |
| last()      | Get the last non-NA value in each column.             | Erhalten des letzten nicht-NA-Werts in jeder Spalte.     |
| mean()      | Calculate the mean (average) of each numeric column. | Berechnen des Durchschnitts (Mittelwert) jeder numerischen Spalte. |
| median()    | Calculate the median (middle value) of each numeric column. | Berechnen des Medians (Mittelwert) jeder numerischen Spalte. |


```python
# Calculate the average age of all passengers
# Berechnen des durchschnittlichen Alters aller Passagiere

average_age = titanic_data["age"].mean()
average_age
```

```python
titanic_data.groupby("sex")["age"].mean()
```

```python
# Calculate the average age by gender
# Berechnen des durchschnittlichen Alters nach Geschlecht

average_age_by_gender = titanic_data.groupby("sex")["age"].mean()
average_age_by_gender
```

```python
# Using a dictionary to specify aggregation methods for different columns
# Verwenden eines Wrterbuchs, um Aggregationsmethoden fr verschiedene Spalten anzugeben

age_stats_by_gender = titanic_data.groupby("sex").agg({"age": ["mean", "median", 'max']})
age_stats_by_gender
```

```python
# Combining multiple aggregations for different columns
# Kombinieren mehrerer Aggregationen fr verschiedene Spalten

combined_stats = titanic_data.groupby("sex").agg({
    "age": ["mean"],
    "survived": "mean"
})
combined_stats
```

## Join Operations

```python
# Sample data for employees
# Beispieldaten fr Mitarbeiter

data_employees = {
    'EmpID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'DeptID': [101, 102, 101, 103]
}
```

```python
# Sample data for departments
# Beispieldaten fr Abteilungen

data_departments = {
    'DeptID': [101, 102, 103, 104],
    'DeptName': ['HR', 'Finance', 'IT', 'Marketing']
}
```

```python
# Creating DataFrames
# Erstellen von DataFrames

df_employees = pd.DataFrame(data_employees)
df_departments = pd.DataFrame(data_departments)
```

```python
# Joining the datasets on DeptID
# Verknpfen der Datenstze anhand von DeptID

merged_df = pd.merge(df_employees, df_departments, on='DeptID')
merged_df
```
