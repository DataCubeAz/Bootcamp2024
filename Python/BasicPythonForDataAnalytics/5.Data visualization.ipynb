## English

- A histogram is a graphical representation of the distribution of data. 
- It is a way to visualize the frequency or count of data points within specific intervals or "bins."
- In this plot, the x-axis represents the data range divided into intervals, 
- while the y-axis shows the frequency or count of data points falling into each interval. 
- Histograms are useful for understanding the data's distribution, identifying patterns, and detecting outliers.

##  Deutsch

- Ein Histogramm ist eine grafische Darstellung der Verteilung von Daten. 
- Es ist eine Möglichkeit, die Häufigkeit oder Anzahl von Datenpunkten innerhalb bestimmter Intervalle oder "Bins" zu visualisieren. 
- In diesem Diagramm repräsentiert die x-Achse den Datenbereich, der in Intervalle unterteilt ist,

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
image = mpimg.imread('Datasets/standard-normal-distribution-example.webp')
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.show()
```

Use the Mean When

- Data is *Normally Distributed*: If your data is symmetrically distributed without skewness, the mean is a good measure of central tendency.
- No Outliers: The mean is sensitive to outliers. If your data doesn’t have outliers, or if they have been removed, using the mean is appropriate.
- Quantitative Data: The mean is suitable for continuous data where values are close to each other.
- while the y-axis shows the frequency or count of data points falling into each interval.
- Histograms are useful for understanding the data's distribution, identifying patterns, and detecting outliers.

Verwenden Sie den Mittelwert, wenn

- Die Daten *normalverteilt* sind: Wenn Ihre Daten symmetrisch verteilt sind ohne Schiefe, ist der Mittelwert ein gutes Maß für die zentrale Tendenz.
- Keine Ausreißer: Der Mittelwert ist empfindlich gegenüber Ausreißern. Wenn Ihre Daten keine Ausreißer haben oder diese entfernt wurden, ist die Verwendung des Mittelwerts angemessen.
- Quantitative Daten: Der Mittelwert eignet sich für kontinuierliche Daten, bei denen die Werte näher beieinander liegen.

```python

ages = df['age'].dropna()
bin_count = 30
plt.figure(figsize=(15, 6))
bin_edges = np.linspace(ages.min(), ages.max(), bin_count + 1)
plt.hist(ages, bins=bin_edges, alpha=0.7, color='blue', edgecolor='black')
plt.axvline(ages.mean(), color='green', linestyle='dashed', linewidth=2)
plt.axvline(np.median(ages), color='orange', linestyle='dashed', linewidth=2)
plt.title('Simulated Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(['Mean', 'Median'])
plt.grid(True)
plt.show();
```

### Boxplot

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
image = mpimg.imread('documents/box_plot.png')
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.show()
```

## English

1. **Bell-Shaped Curve**: This curve shows the normal distribution of our data, indicating that most values are around the center.
2. **Median**: The line in the middle of the pink box is the median, the middle value of our dataset.
3. **Box (IQR)**: The pink box shows the interquartile range (IQR), where the central 50% of our data lies.
4. **Quartiles**: The left edge of the box is the first quartile, and the right edge is the third quartile.
5. **Whiskers**: These lines show the range of the bulk of our data, extending to the furthest points not considered outliers. 
6. **Outliers**: The purple dots are outliers, values that are unusually different from the rest.

## Deutsch

1. **Glockenförmige Kurve**: Diese Kurve zeigt die Normalverteilung unserer Daten und deutet an, dass die meisten Werte um die Mitte liegen.
2. **Median**: Die Linie in der Mitte der rosa Box ist der Median, der mittlere Wert unserer Daten.
3. **Box (IQR)**: Die rosa Box zeigt den Interquartilabstand (IQR), wo die zentralen 50% unserer Daten liegen.
4. **Quartile**: Der linke Rand der Box ist das erste Quartil, und der rechte Rand ist das dritte Quartil.
5. **Whiskers**: Diese Linien zeigen den Bereich des Großteils unserer Daten und erstrecken sich bis zu den am weitesten entfernten Punkten, die nicht als Ausreißer gelten.
6. **Ausreißer**: Die lila Punkte sind Ausreißer, Werte, die ungewöhnlich anders sind als der Rest.


### Find the median , IQR , lower and upper  bound at the age columns 

```python
median=df['age'].median()
percentile_25 = df['age'].quantile(0.25)
percentile_75 =df['age'].quantile(0.75)
```

```python
print("Median:", median)
print("25th Percentile:", percentile_25)
print("75th Percentile:", percentile_75)
```

### IQR (Interquartile Range.)

```python
IQR = percentile_75 - percentile_25

print("Interquartile Range (IQR):")
print(IQR)
```

```python
lower_bound = percentile_25 - 1.5 * IQR
upper_bound = percentile_75 + 1.5 * IQR

print("Lower Bound:")
print(lower_bound)

print("\nUpper Bound:")
print(upper_bound)
```

```python
def plot_numerical_column(dataframe, column):
    """Generates histogram and boxplot for a numerical column."""
    plt.figure(figsize=(18, 10))

    plt.subplot(1, 2, 1)
    sns.histplot(dataframe[column], kde=True)
    plt.title(f"Histogram of {column}")

    plt.subplot(1, 2, 2)
    sns.boxplot(y=dataframe[column])
    plt.title(f"Boxplot of {column}")

    plt.tight_layout()
    plt.show()
```

```python
plot_numerical_column(df, 'age')
```

## English
**Categorical Variables**
Categorical variables are types of data that are divided into groups or categories.

**Nominal Variables**
- These are variables that have two or more categories without any intrinsic order.
- Examples include a person's gender (male, female, other), nationality (American, British, French), or brand of a product (Apple, Samsung, Google).
  
**Ordinal Variables**
- These are categorical variables that have a clear ordered relationship between categories.
- For example, an ordinal variable could be a customer satisfaction rating with categories like "Very Unsatisfied," "Unsatisfied," "Neutral," "Satisfied," and "Very Satisfied," where each category has a clear rank order.
## Deutsch
**Kategoriale Variablen**
Kategoriale Variablen sind Datentypen, die in Gruppen oder Kategorien eingeteilt sind.

**Nominale Variablen**
- Dies sind Variablen mit zwei oder mehr Kategorien, ohne eine inhärente Ordnung.
- Beispiele sind das Geschlecht einer Person (männlich, weiblich, divers), die Nationalität (amerikanisch, britisch, französisch) oder die Marke eines Produktes (Apple, Samsung, Google).
  
**Ordinale Variablen**
- Dies sind kategoriale Variablen mit einer klaren geordneten Beziehung zwischen den Kategorien.
- Zum Beispiel könnte eine ordinale Variable eine Kundenzufriedenheitsbewertung sein mit Kategorien wie "Sehr unzufrieden," "Unzufrieden," "Neutral," "Zufrieden," und "Sehr zufrieden," wobei jede Kategorie eine klare Rangordnung hat.

## Barplot 

```python
outcome_counts = df['survived'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=outcome_counts.index, y=outcome_counts.values)
plt.title('Distribution of Survived Outcomes')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.xticks([0, 1], ['Not survived (0)', 'Survived (1)'])
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(outcome_counts, labels=['Not survived (0)', 'Survived (1)'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Survived Outcomes')
plt.show()
```
