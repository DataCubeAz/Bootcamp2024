##### English
- **Why NumPy?** NumPy makes coding faster and uses less computer memory compared to regular lists. It's perfect for handling lots of numbers or big data sets.
- **Functionality:** With NumPy, you get to use lots of math tools and cool tricks like making operations on a whole list of numbers at once, which you can't do with just Python lists.
- **Community and Ecosystem:** Lots of coders in science and engineering use NumPy because it works really well with other coding tools for data analysis and visualization like Pandas, SciPy, and Matplotlib.
- **Numpy is a fixed-type array:** Unlike lists, where you can mix different types of stuff, NumPy arrays can only have one kind of data. This makes them simpler and faster for calculations.

```python
# Creating a two-dimensional NumPy array with random integers

two_d_array = np.random.randint(5,10, size=(2, 5))
two_d_array
```

```python
two_d_array.ndim
```

```python
two_d_array.shape
```

```python
# Accessing elements using row and column indices
element_0_0 = two_d_array[0, 0]  # First row, first colum
element_1_3 = two_d_array[1, 4]  # Third row, fourth column
element_1_1 = two_d_array[1, 1]  # Second row, second column
print(element_0_0,element_1_1,  element_1_3)
```

```python
two_d_array
```

```python
# Attempting to assign a floating-point number to an integer array

two_d_array[1, 3] = 10.5
```

```python
two_d_array
```

```python
two_d_array.shape
```

```python
# Accessing entire rows or columns using slicing

first_column = two_d_array[0,: ]
first_column
```

## Conditions on Numpy

```python
array_v = np.array([1, 2, 3, 4, 5])
array_v
```

```python
# Performing conditional operations directly on the NumPy array
array_v < 5  # Creates a boolean array
```

```python
array_v[array_v < 5 ]
```

```python
# Filters elements less than 3

elements_less_than_three = array_v[array_v < 3]
elements_less_than_three
```

```python
# Filters elements greater than 3

elements_greater_than_three = array_v[array_v > 3]
elements_greater_than_three
```

```python
# Filters elements not equal to 3
elements_not_equal_three = array_v[array_v != 3]
elements_not_equal_three
```

```python
 # Filters elements equal to 3

elements_equal_three = array_v[array_v == 3]
elements_equal_three
```

```python
 # Filters elements greater than or equal to 3
elements_greater_or_equal_three = array_v[array_v >= 3]
elements_greater_or_equal_three
```
