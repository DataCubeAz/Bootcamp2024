 ## Aggregation with Grouping

| Aggregation | English Description                                   |
|-------------|-------------------------------------------------------|
| count()     | Count the number of non-NA/null entries in each column. |
| sum()       | Calculate the sum of each numeric column.            |
| max()       | Find the maximum value in each column.               |
| min()       | Find the minimum value in each column.               |
| first()     | Get the first non-NA value in each column.            |
| last()      | Get the last non-NA value in each column.             |
| mean()      | Calculate the mean (average) of each numeric column. |




```python
titanic_data.head()
```

```python
# Calculate the average age of all passengers

average_age = titanic_data["person_age"].mean()
average_age
```

```python
# Calculate the average age by gender

average_age_by_gender = titanic_data.groupby("gender")["person_age"].mean()
average_age_by_gender
```

```python
titanic_data.groupby("gender").agg({"person_age" : ["mean", "min", 'max']} )
```

```python
# Using a dictionary to specify aggregation methods for different columns

age_stats_by_gender = titanic_data.groupby("gender").agg({"person_age": ["mean", "min", 'max']})
age_stats_by_gender
```

```python
# Combining multiple aggregations for different columns

combined_stats = titanic_data.groupby("gender").agg({
    "person_age": "max",
    "survived": "mean"
})
combined_stats
```

```python
titanic_data.groupby(["gender", "pclass"]).agg({
    "person_age": "mean",
    "survived": "mean" ,
    'gender': 'count'

})
```

```python
titanic_data.head()
```

```python
# Grouping by multiple columns and calculating different statistics

titanic_data.groupby(["gender", "embarked", "pclass"]).agg({
    "person_age": "mean",
    "survived": "mean" ,
    'gender': 'count'
}).rename(columns={"gender": "count"})

#Port of Embarkation (Cherbourg, Queenstown, Southampton)
```

```python
titanic_data.groupby(["embarked"]).agg({
    "person_age": "mean",
    "survived": "mean" ,
    'gender': 'count'
}).rename(columns={"gender": "count"})
#Port of Embarkation (Cherbourg, Queenstown, Southampton)
```

```python
# Grouping by multiple columns and calculating different statistics

titanic_data.groupby(["gender", "pclass"]).agg({
    "person_age": "mean",
    "survived": "mean",
    'gender' : 'count'
})
```

```python
# We also need the number of person in each group to make a righ analyse

titanic_data.groupby(["gender", "embarked", "pclass"]).agg({
    "person_age": "mean",
    "survived": "mean",
    "gender": "count"}).rename(columns={"gender": "count"})
```

```python
titanic_data.head(3)
```

```python
# Family size analysis

titanic_data["family_size"] = titanic_data["sibsp"] + titanic_data["parch"]
family_size_survival = titanic_data.groupby(["family_size"]).agg({
    "survived":["mean" , 'count']
}).reset_index()
family_size_survival
```

## Join Operations

```python
# Sample data for employees

data_employees = {
    'EmpID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'DeptID': [101, 102, 101, 103]
}
```

```python
# Sample data for departments

data_departments = {
    'DeptID': [101, 102, 103, 104],
    'DeptName': ['HR', 'Finance', 'IT', 'Marketing']
}
```

```python
# Creating DataFrames

df_employees = pd.DataFrame(data_employees)
df_departments = pd.DataFrame(data_departments)
```

```python
df_employees
```

```python
df_departments
```

```python
# Joining the datasets on DeptID

merged_df = pd.merge(df_employees, df_departments, on='DeptID')
merged_df
```
